# Weather Analysis

### Import necessary libraries and import data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import statsmodels

data = pd.read_csv('DailyDelhiClimateTrain.csv')
data.head()
```

### Getting stats on the data

```python
data.describe()
```

```python
data.info()
```

### Visualizing meantemp Over Time

```python
figure = px.line(data, x = 'date', y = 'meantemp', title = 'Mean Temp of Delhi Over the Years')
figure.show()
```

```python
figure = px.line(data, x = 'date', y = 'humidity', title = 'Mean Humidity of Delhi Over the Years')
figure.show()
```

```python
figure = px.line(data, x = 'date', y = 'wind_speed', title = 'Wind Speed of Delhi Over the Years')
figure.show()
```

### Finding relationship between Temp and Humidity

```python
figure = px.scatter(data_frame = data, x="humidity",
                    y="meantemp", size="meantemp", 
                    trendline="ols", 
                    title = "Relationship Between Temperature and Humidity")
figure.show()
```

# Changes in Temp Over Time in Delhi

### Converting date to datetime and adding Year/Month columns

```python
data['date'] = pd.to_datetime(data['date'], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month

data.head()
```

### Temp change over time

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```
